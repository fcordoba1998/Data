## Trabajo Práctico 1 - Probabilidad Discreta
---
---

Simular dos variables aleatorias $X_1$ y $X_2$ que siguen distribuciones de Poisson con parametros $\mu_1 = 1$ y $\mu_2 = 2$ respectivamente. Calcular la variable $Y = X_1 + X_2$. Repetir la simulación 1000 veces. Graficar la distribución de la variable $Y$ y comparar con una distribución de probabilidad adecuada.

---
### **Resolución**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetros de las distribuciones de Poisson
mu1 = 1
mu2 = 2

# Número de simulaciones
num_simulations = 1000

# Simulación de las variables aleatorias X1 y X2
X1 = np.random.poisson(mu1, num_simulations)
X2 = np.random.poisson(mu2, num_simulations)

# Cálculo de la variable Y = X1 + X2
Y = X1 + X2

# Histograma de la variable Y
plt.hist(Y, bins=range(0, max(Y) + 1), density=True, alpha=0.5, label='Simulación')

# Distribución de probabilidad adecuada (Poisson)
y_values = np.arange(0, max(Y) + 1)
poisson_probs = poisson.pmf(y_values, mu=mu1 + mu2)
plt.plot(y_values, poisson_probs, 'bo-', label='Poisson Teórico')

plt.xlabel('Valor de Y')
plt.ylabel('Probabilidad')
plt.title('Distribución de la variable Y')
plt.legend()
plt.show()
````

   ```python

   ```

   Aquí se definen los parámetros μ1 y μ2 para las distribuciones de Poisson de X1 y X2, respectivamente.

3. **Número de simulaciones**:


   Esto establece cuántas veces se realizará la simulación.

