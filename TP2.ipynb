Considerar 20 variables aleatorias que siguen una distribución normal estándar $(X_1, . . . , X_{20})$ y hacer el cambio de variables $Y_i = X^2_i$. Identificar que función de densidad de probabilidad siguen las nuevas variables $Y_i$. Construir una nueva variable aleatoria $Z=\sum^{20}_{i=1} Y_i$. A continuación simular las 20 variables $X_i$ y calcular el valor $Z$ correspondiente. Repetir este proceso 10.000 veces y hacer un histograma de frecuencias de $Z$. Comparar el histograma con una distribucion chi-cuadrado y otra normal con parametros apropiados.

---
### **Resolución**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm

# Número de simulaciones
num_simulations = 10000

# Número de variables aleatorias
num_variables = 20

# Crear una matriz para almacenar los resultados de Z en cada simulación
results_Z = np.zeros(num_simulations)

# Realizar las simulaciones
for i in range(num_simulations):
    # Generar 20 variables aleatorias distribuidas normalmente
    X = np.random.normal(0, 1, num_variables)

    # Calcular las variables Yi = X2i
    Yi = X ** 2

    # Calcular Z = ∑20i=1 Yi
    Z = np.sum(Yi)

    # Almacenar el resultado en la matriz
    results_Z[i] = Z

# Hacer un histograma de frecuencias de Z
plt.hist(results_Z, bins=50, density=True, alpha=0.6, color='b', label='Histograma de Z')

# Calcular la distribución chi-cuadrado con 20 grados de libertad
x = np.linspace(0, 40, 400)
chi2_dist = chi2.pdf(x, df=num_variables)

# Calcular la distribución normal con parámetros apropiados
mean_Z = num_variables  # Media de Z
std_Z = np.sqrt(2 * num_variables)  # Desviación estándar de Z
normal_dist = norm.pdf(x, loc=mean_Z, scale=std_Z)

# Graficar las distribuciones teóricas
plt.plot(x, chi2_dist, 'r--', label='Chi-cuadrado (k=20)')
plt.plot(x, normal_dist, 'g--', label='Normal')

# Configurar el gráfico
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de probabilidad')
plt.title('Histograma de frecuencias de Z y comparación con distribuciones teóricas')
plt.legend()

# Mostrar el gráfico
plt.show()
````


 el problema implica la generación de variables aleatorias, transformaciones y comparaciones con distribuciones teóricas para analizar la distribución de la variable aleatoria Z, que es la suma de los cuadrados de variables normales estándar. Se utiliza la distribución chi-cuadrado con 20 grados de libertad y una aproximación normal para este análisis.

Dado que tenemos 20 variables aleatorias (X1, X2, ..., X20) que siguen una distribución normal estándar (con media μ=0 y desviación estándar σ=1), se nos pide realizar varias tareas:

1. **Transformación de Variables (Yi = Xi^2)**:
   - Primero, aplicamos una transformación a cada una de las variables Xi para obtener las nuevas variables (Y1, Y2, ..., Y20) utilizando la relación Yi = Xi^2. Esto significa que cada Yi representa el cuadrado de su correspondiente Xi. Como resultado, las variables Yi siguen una distribución chi-cuadrado con 1 grado de libertad (k=1), ya que el cuadrado de una variable normal estándar sigue esta distribución.

2. **Construcción de la Variable Aleatoria Z**:
   - Luego, construimos una nueva variable aleatoria Z como la suma de las 20 variables Yi: Z = Y1 + Y2 + ... + Y20. La suma de variables chi-cuadrado independientes con k grados de libertad sigue una distribución chi-cuadrado con k grados de libertad. En este caso, Z sigue una distribución chi-cuadrado con 20 grados de libertad (k=20), ya que estamos sumando 20 variables chi-cuadrado con 1 grado de libertad.

3. **Simulación de Variables y Histograma**:
   - Luego, realizamos un proceso de simulación donde generamos 10,000 conjuntos de 20 variables Xi y calculamos la correspondiente variable Z en cada conjunto siguiendo las relaciones anteriores. Esto nos permite obtener una muestra de 10,000 valores de Z.

4. **Histograma de Frecuencias de Z**:
   - Finalmente, construimos un histograma de frecuencias de los valores de Z. Un histograma es una representación gráfica de la distribución de datos que muestra cuántas veces caen los valores en intervalos específicos. En este caso, estamos interesados en observar cómo se distribuyen los valores de Z en los diferentes intervalos.

Comparación con Distribuciones Teóricas:
- Comparamos el histograma resultante con dos distribuciones teóricas:
  - Distribución Chi-cuadrado: Sabemos que Z sigue una distribución chi-cuadrado con 20 grados de libertad (k=20). Por lo tanto, comparamos el histograma con la función de densidad de probabilidad de la distribución chi-cuadrado con 20 grados de libertad.
  - Distribución Normal (Aproximación): Cuando el parámetro k de la distribución chi-cuadrado es mucho mayor que 1, podemos usar una distribución normal como una aproximación. En este caso, comparamos el histograma con una función de densidad de probabilidad de una distribución normal con media μ=20 y desviación estándar σ=√(2*20), que es una aproximación válida según la ley de los grandes números.
